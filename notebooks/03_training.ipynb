Training is super simple, all you need to do is import train from src/train
and then give your model a name. If you want to change any of the parameters you
can by giving them a new value. If you want to change the length of the max title length
you will need to go to 02_preprocessing.ipynb and change it there too.
___
You may also need to change the working directory of the notebook to the MIND_Recommender folder. You can do this by typing **%cd ***path to your folder***** at the top of the cell you are running.
